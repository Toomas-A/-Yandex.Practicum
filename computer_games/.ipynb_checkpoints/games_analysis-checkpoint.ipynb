## Шаг 2. Подготовка данных

- Заменим названия столбцов.

```python
data.set_axis(['name','platform', 'year_of_release', 'genre', 'am_sales', 'eu_sales', 
'jp_sales', 'other_sales', 'сritic_score', 'user_score', 'rating'], axis = 'columns', inplace=True)
data.head()
```

```python
data['name'].value_counts().head()
```

Так как в наших данных присутствуют строки с одинаковым названием игр воспользуемся этим для заполнения отсутствующих данных на основании годов выпуска игр с одним и тем же названием, выпущенных для разных платформ. Заменим NaN средним значением.

```python
data['year_of_release'] = data.groupby(data['name'])['year_of_release'].transform(lambda x: x.fillna(x.mean()))
data.info()
```

Осталось 146 игр, которые не имеют даты выпуска. Так как это уже незначительная часть и хотелось бы некоторой вариативности заполнения, заменим NaN на предыдущие значения (установив предел замены после двух успешных). 

И заменим десятичные значения года выпуска на целые.

```python
data['year_of_release'].fillna(method='ffill', limit = 2, inplace=True)
data.info()
data['year_of_release'] = data['year_of_release'].astype(int)
```

- Оценки одной и той же игры на разных платформах отличаются, но не кардинально, поэтому заполним отсутствующие значения в столбцах `сritic_score` и `user_score`, заменив их на среднее значение оценок одной и той же игры для разных платформ. Остальные оставим как есть. 

>Таким образом заполнили *961* пропуск в `сritic_score` и *429* пропусков в `user_score`.

- В столбце `user_score` 2424 строки с значением "tbd", что означает "to be defined". Проанализировав оценочные условия различных платформ, пришли к выводу, что "tbd" - это продукт без оценки, т.к. игра либо [новая](https://www.metacritic.com/game/pc/doom-eternal/critic-reviews), либо [непопулярная](https://www.metacritic.com/game/pc/haunted-house-cryptic-graves/critic-reviews) и не набрала определённого количества оценок, т.н. момент ["калибровки"](https://www.metacritic.com/faq#item13). Исходя из этого заменим значения "tbd" на 0.

- Преобразуем данные в столбце `user_score` в числовые.

```python
data['сritic_score'] = data.groupby(data['name'])['сritic_score'].transform(lambda x: x.fillna(x.mean()))
data[data['сritic_score'].isnull()]
```

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = 0
data['user_score'] = [float(item) for item in data['user_score']]
data['user_score'] = data.groupby(data['name'])['user_score'].transform(lambda x: x.fillna(x.mean()))
```

---

- Добавим столбец с суммарной продажей игры во всех регионах.

```python
data['total_sales'] = data.loc[:,'am_sales':'other_sales'].sum(axis=1)
data.head()
```

### Проанализируем, как менялись продажи по платформам. 

- Так как в наших данных представлены платформы вышедшие на рынок в достаточно большой промежуток времени (в разные десятилетия и даже века))), сгруппируем их для наглядности по десятилетиям. 
- Визуализируем глобальные продажи по платформам.
- Выберем актуальные платформы (для 2ой декады 21 века) с наибольшими суммарными продажами и построем распределение по годам.

```python
data['decade'] = 10 * (data['year_of_release'] // 10) # добавим столбец с группировкой года выхода платформы по декадам
platf_rev = data.pivot_table('total_sales', index='platform', columns='decade', aggfunc='sum', fill_value=0) # платформы по декадам
platf_rev['total_rev'] = platf_rev.loc[:,'1980':'2010'].sum(axis=1)
platf_rev.reset_index(inplace = True)
platf_rev.columns = ['platform', '1980', '1990', '2000', '2010', 'total_rev']
platf_rev.sort_values(by='total_rev')

platf_rev.loc['total'] = platf_rev.sum(numeric_only=True) # игого по декадам
platf_max = platf_rev[platf_rev['2010'] > 0].sort_values('total_rev') # с продажами в 2010-х годах
                      
platf_max
```

Хорошо видно развитие платформ по динамике выручки. Например, пик продаж [Atari 2600](https://ru.wikipedia.org/wiki/Atari_2600) пришелся на начало 80-х с последующим закатом платформы. 

> Кстати, Atari 2600 самая долгоживущая платформа в истории видеоигр — 14 лет и 2 месяца (1 января 1992 года прекратилась официальная поддержка).

Или, например, развитие проекта Play Station - PS, PS2, PS3. Представленная в 2013 году компанией Sony консоль PS4, объём продаж которой по нашим данным, за неполные 3 года превысил 300 млн.долларов.

> Показатели продаж PS4 аналитики считают [самым успешным запуском](https://www.vg247.com/2013/12/12/ps4-is-worlds-biggest-console-launch-analyst/) в истории игровых консолей.

### Рассмотрим выделенные платформы более детально

#### PS4

Определим игры приносящие максимальную ежегодную выручку с момента появления платформы. 

```python
ps4_data = data.query('platform == "PS4"')
ps4_games = ps4_data.groupby(['name', 'year_of_release']).sum()[['am_sales', 'eu_sales', 'jp_sales', 'other_sales', 'total_sales']]
ps4_games = ps4_games.reset_index(inplace = False)
ps4_games.sort_values(by = 'total_sales', ascending = False).head(10)
```

Для наглядности разброса ежегодной глобальной выручки построим график «ящик с усами» по играм платформы.

```python
figure(num=None, figsize=(6, 6), dpi=80, facecolor='w')

sns.boxplot('year_of_release', 'total_sales', data=ps4_games)
```

Основные продажи приходятся на десяток игр. Выручка с каждым годом падает. 

Связано это с изменением модели монетизации игровой индустрии. [Рынок мобильных игр развивается настолько быстро, что по прогнозам Newzoo к 2021 году превратится в индустрию с доходом в сто миллиардов долларов, в то время как прибыль от обычных видеоигр кое-как перевалит за тридцать два миллиарда](https://dtf.ru/mobile/50025-industriya-v-karmane-kak-ustroen-mir-mobilnyh-igr). Если в 1990-х годах основным легальным способом обзавестись игрой была покупка классической «коробки» и консоли, то с развитием мобильных гаджетов массовый рынок начал мигрировать на мобильные устройства (смартфоны и планшеты). Плюс развивается модель  free-to-play (f2p), когда сама игра бесплатна (монетизация происходит за счет внутриигровых покупок).

Проанализируем топ игры данной платформы. Чтобы сконцентрировать внимание на приносящих наибольшую выручку играх, сделаем отсечку (исходя из графика «ящик с усами») на 1 млн. долларов. И ограничим наблюдения периодом с 2014г.

## Шаг 5. Проверка гипотез:

### 1) Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Есть 2 генеральных совокупности - пользовательские рейтинги платформ `XOne` и `PC`. Чтобы проверить гипотезу, различается ли пользовательские рейтинги этих платформ, применим метод scipy.stats.ttest_ind(). За нулевую гипотезу примем, что рейтинги НЕ различаются (одинаковые). Альтернативная гипотеза - рейтинги разные. Границу нулевой гипотезы определим в [1%](https://www.statmethods.ru/stati/vybor-urovnya-znachimosti-pri-proverke-statisticheskikh-gipotez/). Доверительный интервал - 99%.

```python
xone_score = data.query('platform == "XOne"')[['user_score']]
pc_score = data.query('platform == "PC"')[['user_score']]

xone_score_mean = xone_score['user_score'].mean()
xone_score_mean
print('Средний пользовательский рейтинг платформы Xbox One:', xone_score_mean)

pc_score_mean = pc_score['user_score'].mean()
print('Средний пользовательский рейтинг платформы PC:', pc_score_mean)
print('\n')

xone_score = xone_score['user_score'].dropna()
pc_score = pc_score['user_score'].dropna()
alpha = .01 # критический уровень статистической значимости

res_score_dif = st.ttest_ind(xone_score, pc_score) 
    
print('p-value:', res_score_dif.pvalue)

if (res_score_dif.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Принимаем нулевую гипотезу. Дополнительным подтверждением того, что рейтинги похожи может служить неначительная разница между средними значениями рейтингов этих платформ. Гипотеза подтверждена, рейтинги одинаковые.

Как подтверждение правильности нашего вывода, определим разброс/стандартное отклонение и вероятность того, что оценка попадёт в эту область. Для чего посчитаем величину дисперсии и стандартного отклонения. Построим гистограмму посльзовательских рейтингов.

#### Xbox One

```python
var_xone = np.var(xone_score, ddof=1)
dev_xone = np.sqrt(var_xone)
print('Дисперсия равна:', var_xone)
print('Стандартное отклонение равно:', dev_xone)
```

```python
distr_xone = st.norm(xone_score_mean, dev_xone) 

x1_xone = xone_score_mean - dev_xone
x2_xone = xone_score_mean + dev_xone

prob_xone = distr_xone.cdf(x2_xone) - distr_xone.cdf(x1_xone)
print('Вероятность оценки платформы Xbox One от {:.0f} баллов до {:.0f} баллов равна {:.0%}.'.format(x1_xone, x2_xone, prob_xone))
```

#### PC  
То же сделаем для платформы PC.

```python
var_pc = np.var(pc_score, ddof=1)
dev_pc = np.sqrt(var_pc)
print('Дисперсия равна:', var_pc)
print('Стандартное отклонение равно:', dev_pc)
```

```python
distr_pc = st.norm(pc_score_mean, dev_pc) 

x1_pc = pc_score_mean - dev_pc
x2_pc = pc_score_mean + dev_pc

prob_pc = distr_pc.cdf(x2_pc) - distr_pc.cdf(x1_pc)
print('Вероятность оценки платформы PC от {:.0f} баллов до {:.0f} баллов равна {:.0%}.'.format(x1_pc, x2_pc, prob_pc))
```

```python
xone_score.hist(bins=30,alpha=1, range=(1, 10), color = 'red')
pc_score.hist(bins=30,alpha=0.5, range=(1, 10), color = 'grey')
plt.legend(['XOne', 'PC'])
```

Хотя количество рейтингов предсказуемо отличается (XBox One поступила в продажу в конце 2013г., в то время как игры на PC выходят десятилетиями), средние пользовательские рейтинги похожи. Нормальное распределение одного графика полностью перекрывается другим (нулевая гипотеза подтверждена - рейтинги одинаковые).

### 2) Средние пользовательские рейтинги жанров Action и Sports разные.

Есть 2 генеральных совокупности - пользовательские рейтинги жанров Action и Sports. Чтобы проверить гипотезу, различается ли пользовательские рейтинги этих платформ, применим метод scipy.stats.ttest_ind(). За нулевую гипотезу примем, что рейтинги НЕ различаются (одинаковые). Альтернативная гипотеза - рейтинги разные. Границу нулевой гипотезы определим в 5%. Доверительный интервал - 95%.

```python
action_score = data.query('genre == "Action"')[['user_score']]
sports_score = data.query('genre == "Sports"')[['user_score']]
# action_score['user_score']
# sports_score['user_score'].dropna()
action_score_mean = action_score['user_score'].mean()

print('Средний пользовательский рейтинг жанра Action:', action_score_mean)

sports_score_mean = sports_score['user_score'].mean()
print('Средний пользовательский рейтинг жанра Sports:', sports_score_mean)
print('\n')

action_score = action_score['user_score'].dropna()
sports_score = sports_score['user_score'].dropna()

alpha = .05 # критический уровень статистической значимости

res_genre_dif = st.ttest_ind(action_score, sports_score) 
    
print('p-value:', res_genre_dif.pvalue)

if (res_genre_dif.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Отвергаем нулевую гипотезу. Даже если сравнить средние величины, то они различаются более чем на 13%. Гипотеза о том, что средние пользовательские рейтинги жанров Action и Sports разные подтверждена.

От «золотой эры аркадных игр» (начало 80-х), когда стоимость и сроки разработки была минимальной, поэтому и количество игр с каждым годом росло. До момента, когда миниатюризация аппаратного обеспечения и массовое распространение мобильных телефонов способствовало появлению мобильных игровых приложений и облачных платформ. С параллельным развитием компьютерных технологий, увеличилась стоимость и сроки разработки, так как увеличивающаяся сложность графики и программирования в целом требовала все большего штата специалистов. Большинство современных компьютерных игр разрабатываются в течение от одного до трех лет, что соответственно создает повышенные требования к бюджету игры ([стоимость разработки может достигать десятков миллионов долларов](https://habr.com/ru/company/mailru/blog/347474/)).
В нашем исследовании виден рассвет и закат платформ, цикличность развития рынка и динамика продаж.

## Выводы

1)	Рынок игр поделён между несколькими основными компаниями-разработчиками - Sony (PS), Nintendo (Wii, DS, GBA), Microsoft (X360, XB).

2)	Платформы развивается и умирает в среднем в течении 5-8 лет.

3)	По состоянию на конец 2016г. можно констатировать падение продаж по всем платформам. Возможно это связано с развитием на тот момент мобильных игр и переориентированием рынка в облачные сервисы. ![](https://hsto.org/webt/9s/z-/vk/9sz-vkludehsnte_ooolirw6viw.jpeg)

4)	При планировании, следует обращать внимание на платформы, которые только появляются, либо находятся "на взлёте" - 2-3 года с момента выпуска.

5)	При планировании рекламного бюджета на 2017г. стоит обратить внимание на следующие платформы: PS4, WiiU, XOne, PC.

6)	На сегодняшний день основными рынками сбыта (приносящие максимальную выручку) являются Северная Америка и Европа. Необходимо расширять азиатский рынок.

7)	Одна и та же игра может быть популярна на одной платформе и не популярна на другой, соответственно приносит разную выручку.

8)	У каждой платформы есть несколько самых популярных игр (лидеры продаж), приносящих максимальную выручку. 

9)	Самые продаваемые игры в глобальном масштабе следующих жанров (впорядке убывания): Action, Shooter, Sports.

10)	Наименьший объем выручки приносят игры жанров Puzzle и Stretegy.

11)	Взаимосвязь между оценками критиков/пользователей и уровнем продаж отсутствует.
